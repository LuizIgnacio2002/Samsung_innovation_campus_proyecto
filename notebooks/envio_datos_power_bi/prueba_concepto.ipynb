```python
import pandas as pd
import os
import re

# Ruta del archivo CSV
csv_path = '../data/03_casos/df_52_cleaned.csv'

# Leer el CSV
df = pd.read_csv(csv_path)

# Obtener solo el nombre del archivo (sin ruta ni extensión)
file_name = os.path.splitext(os.path.basename(csv_path))[0]  # 'df_52_cleaned'

# Extraer el número entre los guiones bajos usando regex
match = re.search(r'_(\d+)_', file_name)
if match:
    sub_reg_nt = match.group(1)  # '52'
else:
    sub_reg_nt = None  # Por si no hay coincidencia

# Añadir la columna con el número extraído
df['sub_reg_nt'] = sub_reg_nt
```

```python
df.head()
```

```python
df = df[[
    'ano',
    'semana',
    'sub_reg_nt',
    'hospitalizados_men5',
]]
```

```python
import pandas as pd
import requests
import json
import os
import re

# Crear lista de diccionarios para Power BI
payload = []
for _, row in df.iterrows():
    payload.append({
        "ano": str(row['ano']),
        "semana": str(row['semana']),
        "sub_reg_nt": str(row['sub_reg_nt']),
        "real_hospitalizados_men5": float(row['hospitalizados_men5']),
        "pred_hospitalizados_men5":  None   # sin valor por ahora
    })

# === 4️⃣ Enviar al endpoint de Power BI ===
endpoint = "https://api.powerbi.com/beta/37e0b05d-41da-4172-9c5f-b0fc5e15667e/datasets/05e1d5d6-02f7-412e-aa80-7e526fcc309c/rows?experience=power-bi&key=NRfvRQepSIDZiBChXLVBmqGgpwnotL8frml%2FRNbYGw5lwTgC9y1atrgi2jz91KHdi9Ha8bAPtCSeZrZ5GOkxdg%3D%3D"

headers = {
    "Content-Type": "application/json"
}

# Enviar los datos
response = requests.post(endpoint, data=json.dumps(payload), headers=headers)

# === 5️⃣ Mostrar resultado ===
if response.status_code == 200:
    print(f"✅ Datos enviados correctamente ({len(payload)} filas) al dataset de Power BI.")
else:
    print(f"❌ Error al enviar datos. Código: {response.status_code}")
    print("Detalle del error:", response.text)
```
